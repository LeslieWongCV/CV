```python
import numpy as np
import cv2
import scipy.ndimage as nd
import matplotlib.pyplot as plt
LENA = cv2.imread('/Users/leslie/Desktop/-/LENA.jpg', 0)


LENA_f = np.copy(LENA)
LENA_f_ = LENA_f.astype('float')

Result = nd.gaussian_laplace(LENA_f_, sigma=1)   #scipy.ndimage 

plt.imshow(Result,cmap=plt.get_cmap('gray'))
plt.show()
```
